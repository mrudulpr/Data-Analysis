```python
def detect_outlliers(data):
    list =[]
    data = sorted(data)
    q1 =  np.percentile(data,25)
    q3 = np.percentile(data,75)

    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 - 1.5 * IQR

    for i in data:
        if(i < lower_bound) or (i > upper_bound):
            list.append(i)
    return list
```

```python
def outlier_handler(data):
    tenth_per = np.percentile(data,10)
    nintenth_per = np.percentile(data,90)

    b = np.where(data < tenth_per , tenth_per, data)
    b1 = np.where(b > nintenth_per, nintenth_per, b)
    return b1
 
```

```python
outlier = ['total_rooms','total_bedrooms','population','households','median_income','median_house_value']
for i in outlier:
    out = detect_outlliers(data[i])
    print('outliers in ' ,i, 'attribute' , out)
```

```python
outlier = ['total_rooms','total_bedrooms','population','households','median_income','median_house_value']

for i in outlier:
    data[i] = outlier_handler(data[i])
```
